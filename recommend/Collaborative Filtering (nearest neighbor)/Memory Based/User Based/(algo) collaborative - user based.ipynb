# 데이터 관리
1000-2000등의 정보를 담은 데이터프레임을 생성하고 csv로 저장하여 관리합니다.

추천 서비스를 동작할 때 이 데이터를 활용합니다.

```
import glob
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import pairwise_distances
```

```
path = r'./userdetail_1000_to_2000'
filenames = glob.glob(path+"/*.csv")

userList = []
for filename in filenames:

    user = pd.read_csv(filename, index_col=None, header=0)
    user['userID'] = filename[filename.rfind('/')+1:-4]
    userList.append(user)

users = pd.concat(userList, ignore_index=True)
```

```
# users
```

```
def func(row):
    if(row['정답여부'] == True):
        return 1.0
    elif(row['정답여부'] == False):
        return 0.5
    else:
        return 0

users['정답여부'] = users.apply(func, axis=1)
```

```
# users
```

```
users.to_csv('./userdetail_1000_to_2000.csv', index = False)
```

```
def find_n_neighbours(df,n):
    order = np.argsort(df.values, axis=1)[:, :n]
    df = df.apply(lambda x: pd.Series(x.sort_values(ascending=False)
           .iloc[:n].index, 
          index=['top{}'.format(i) for i in range(1, n+1)]), axis=1)
    return df
```

```
# top 30 neighbours for each user
sim_user_30 = find_n_neighbours(similarity,30)
```
